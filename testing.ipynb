```python
# import torch
# import numpy as np
# import matplotlib.pyplot as plt
# from model.Networks import unet
# from matplotlib.colors import ListedColormap

# from dataset.landslide_dataset import LandslideDataSet
# from torch.utils.data import DataLoader

# # Load the model and set it to evaluation mode
# model = unet(n_classes=2)
# model.load_state_dict(torch.load('./snapshots/unet_trained_model.pth'))
# model.eval()

# # Load test data
# test_dataset = LandslideDataSet('./dataset/', './dataset/test.txt', set='unlabeled')
# test_loader = DataLoader(dataset=test_dataset, batch_size=1, shuffle=False)
# cmap = ListedColormap(["purple", "yellow"])

# # Perform inference and visualize
# for index, batch in enumerate(test_loader):  
#     if index >= 10:
#         break
#     image, _, name = batch
#     image = image.float()  # No need to move to GPU
#     name = name[0].split('.')[0].split('/')[-1].replace('image','mask')
#     print(index+1, '/', len(test_loader), ': Testing ', name)  
#     with torch.no_grad():
#         pred = model(image)
    
#     # Process predictions
#     pred_class = np.argmax(pred.squeeze().numpy(), axis=0)
    
#     # Plot original image and prediction
#     plt.figure(figsize=(12, 6))
#     plt.subplot(1, 2, 1)
#     plt.title(f"Original Image {index}")
#     plt.imshow(image.squeeze().numpy()[0],cmap='gist_earth')
# #     plt.imshow(image.squeeze().numpy()[0], cmap='gray')

#     plt.subplot(1, 2, 2)
#     plt.title(f"Predicted Mask {index}")
#     plt.imshow(pred_class,cmap=cmap)
#     plt.show()
    
#     # Postprocess prediction to get binary matrix
#     pred = pred.squeeze().numpy()
#     pred_class = np.argmax(pred, axis=0)  # Get the class with the highest score for each pixel
#     binary_matrix = (pred_class == 1).astype(int)  # Convert landslide class to 1, non-landslide to 0
    
#     print(f"Binary Mask for Image {index} ({name[0]}):")
#     print(binary_matrix)
#     print("\n")
    
#     total_pixels = pred_class.size
#     landslide_pixels = np.sum(pred_class == 1)
#     non_landslide_pixels = np.sum(pred_class == 0)
    
#     landslide_percentage = (landslide_pixels / total_pixels) * 100
#     non_landslide_percentage = (non_landslide_pixels / total_pixels) * 100

#     # Print results
#     print(f"Image {index + 1}:")
#     print(f"Landslide area: {landslide_percentage:.2f}%")
#     print(f"Non-landslide area: {non_landslide_percentage:.2f}%\n")
import torch
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import random
from model.Networks import unet
from dataset.landslide_dataset import LandslideDataSet
from torch.utils.data import DataLoader

# Load the model and set it to evaluation mode
model = unet(n_classes=2)
model.load_state_dict(torch.load('./snapshots/unet_trained_model.pth'))
model.eval()

# Load test data
test_dataset = LandslideDataSet('./dataset/', './dataset/test.txt', set='unlabeled')
test_loader = DataLoader(dataset=test_dataset, batch_size=1, shuffle=False)

# Randomly select 10 indices from the test dataset
random_indices = random.sample(range(len(test_loader)), 10)
cmap = ListedColormap(["purple", "yellow"])
# Perform inference and visualize
for index, i in enumerate(random_indices):
    batch = test_loader.dataset.__getitem__(i)  # Manually get the batch from the dataset
    image, _, name = batch
    image = torch.tensor(image).float().unsqueeze(0)  # Convert NumPy array to PyTorch tensor, then to float, and add batch dimension
    name = name[0].split('.')[0].split('/')[-1].replace('image', 'mask')
    print(index + 1, '/', 10, ': Testing ', name)  
    
    with torch.no_grad():
        pred = model(image)
    
    # Process predictions
    pred_class = np.argmax(pred.squeeze().numpy(), axis=0)
    
    # Plot original image and prediction
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.title(f"Original Image {index}")
    plt.imshow(image.squeeze().numpy()[0], cmap='gist_earth')

    plt.subplot(1, 2, 2)
    plt.title(f"Predicted Mask {index}")
    plt.imshow(pred_class, cmap=cmap)
    plt.show()
    
    # Postprocess prediction to get binary matrix
    pred = pred.squeeze().numpy()
    pred_class = np.argmax(pred, axis=0)  # Get the class with the highest score for each pixel
    binary_matrix = (pred_class == 1).astype(int)  # Convert landslide class to 1, non-landslide to 0
    
    print(f"Binary Mask for Image {index} ({name[0]}):")
    print(binary_matrix)
    print("\n")
    
    total_pixels = pred_class.size
    landslide_pixels = np.sum(pred_class == 1)
    non_landslide_pixels = np.sum(pred_class == 0)
    
    landslide_percentage = (landslide_pixels / total_pixels) * 100
    non_landslide_percentage = (non_landslide_pixels / total_pixels) * 100

    # Print results
    print(f"Image {index + 1}:")
    print(f"Landslide area: {landslide_percentage:.2f}%")
    print(f"Non-landslide area: {non_landslide_percentage:.2f}%\n")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# List to store information of images with landslide area > 98% and highest landslide area
selected_images = []

# Perform inference and analyze landslide percentages
for index, (image, size, name) in enumerate(test_loader):
    if index>=1000:
        break
    image = image.float()
    with torch.no_grad():
        pred = model(image)
    
    # Process predictions
    pred_class = np.argmax(pred.squeeze().numpy(), axis=0)
    
    # Calculate percentage of landslide and non-landslide pixels
    total_pixels = pred_class.size
    landslide_pixels = np.sum(pred_class == 1)
    non_landslide_pixels = np.sum(pred_class == 0)
    
    landslide_percentage = (landslide_pixels / total_pixels) * 100
    non_landslide_percentage = (non_landslide_pixels / total_pixels) * 100

    # Check if landslide percentage is above 98%
    if landslide_percentage > 98:
        selected_images.append({
            "index": index + 1,
            "landslide_percentage": landslide_percentage,
            "non_landslide_percentage": non_landslide_percentage,
            "image": image,
            "pred_class": pred_class
        })

# Sort images by maximum landslide percentage and select top 5
selected_images = sorted(selected_images, key=lambda x: -x["landslide_percentage"])[:5]

# Display results for the top 5 selected images
for image_info in selected_images:
    print(f"Image {image_info['index']}:")
    print(f"Landslide area: {image_info['landslide_percentage']:.2f}%")
    print(f"Non-landslide area: {image_info['non_landslide_percentage']:.2f}%\n")

    # Plot original image and prediction
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.title(f"Original Image {image_info['index']}")
    plt.imshow(image_info['image'].squeeze().numpy()[0], cmap='gist_earth')

    plt.subplot(1, 2, 2)
    plt.title(f"Predicted Mask {image_info['index']}")
    plt.imshow(image_info['pred_class'], cmap='binary')
    plt.show()
```
